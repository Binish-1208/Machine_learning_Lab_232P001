```python
# Load the Iris dataset
from sklearn.datasets import load_iris

iris = load_iris()
print(iris)
```

```python
# Convert dataset to a Pandas DataFrame for easy manipulation
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
# Add the target column (class labels: 0, 1, 2)
df['target'] = iris.target
# Map target values to actual flower names (e.g., 0 → setosa)
df['flower_name'] = df['target'].apply(lambda x: iris.target_names[x])
df.target
```

```python
# Function to plot any 2 dimensions for two flower types
def plot_dimensions(x, y, xlabel, ylabel, title):
    plt.title(f"By Binish Moosa\n{title}") # Title with name
    plt.xlabel(xlabel) # X-axis label
    plt.ylabel(ylabel) # Y-axis label
    # Plot Setosa data points in red with '+' marker
    plt.scatter(df0[x], df0[y], color="red", marker='+', label="Setosa")
    # Plot Versicolor data points in blue with '.' marker
    plt.scatter(df1[x], df1[y], color="blue", marker='.', label="Versicolor")
    plt.legend() # Show legend
    plt.show() # Display plot
plot_dimensions('sepal length (cm)', 'sepal width (cm)',
'Sepal Length', 'Sepal Width',
"Plot Sepal Length vs Sepal Width")
# Plot Petal dimensions for Setosa and Versicolor
plot_dimensions('petal length (cm)', 'petal width (cm)',
'Petal Length', 'Petal Width',
"Plot Petal Length vs Petal Width")    
```

```python
# Prepare features and target variable
# Drop 'target' and 'flower_name' columns to keep only numeric features
X = df.drop(['target', 'flower_name'], axis='columns')
# Target labels (0, 1, 2 corresponding to the three flower classes)
y = df['target']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
models = {
'default': SVC(), # Default SVM with RBF kernel
'C=1': SVC(C=1), # Regularization parameter C = 1
'C=10': SVC(C=10), # Regularization parameter C = 10 (less
'gamma=10': SVC(gamma=10), # Set gamma = 10 (controls influence of
'linear': SVC(kernel='linear') # Use linear kernel instead of RBF
}
```

```python
for name, model in models.items():
    model.fit(X_train, y_train) # Fit model to training data
```

```python
prediction = models['default'].predict([[4.8, 3.0, 1.5, 0.3]])
# Print predicted flower name using the target name mapping
print(f"Prediction for input [4.8, 3.0, 1.5, 0.3]: {iris.target_names[prediction[0]]}")
```

```python
# SVM using iris dataset with GUI --> for students

import tkinter as tk
from tkinter import messagebox
from sklearn.datasets import load_iris
from sklearn.svm import SVC

# Load Iris data and train SVM model
iris = load_iris()
X, y = iris.data, iris.target
model = SVC()
model.fit(X, y)

# GUI prediction function
def predict():
    try:
        sl = float(entry1.get())
        sw = float(entry2.get())
        pl = float(entry3.get())
        pw = float(entry4.get())

        # Predict using trained model
        prediction = model.predict([[sl, sw, pl, pw]])[0]
        flower = iris.target_names[prediction]

        # Show result
        messagebox.showinfo("Prediction Result", f"Predicted Flower: {flower}")

    except ValueError:
        messagebox.showerror("Input Error", "Please enter valid numeric values!")

# Create main window
root = tk.Tk()
root.title("Iris Flower Predictor")
root.geometry("380x280")

# Light pink theme
bg_color = "#ffe6f2"   # soft pink
btn_color = "#ff99cc"  # medium pink
btn_hover = "#ff66b2"  # darker pink
text_color = "#4a004a" # deep purple

root.config(bg=bg_color)

# Input labels + entries
labels = ["Sepal Length (cm):", "Sepal Width (cm):", "Petal Length (cm):", "Petal Width (cm):"]
entries = []

for i, text in enumerate(labels):
    tk.Label(root, text=text, bg=bg_color, fg=text_color, font=("Arial", 11, "bold")).grid(row=i, column=0, padx=10, pady=5, sticky="w")
    entry = tk.Entry(root, bg="white", fg=text_color, justify="center", relief="solid")
    entry.grid(row=i, column=1, pady=5)
    entries.append(entry)

entry1, entry2, entry3, entry4 = entries

# Hover effect for button
def on_enter(e):
    predict_btn.config(bg=btn_hover)

def on_leave(e):
    predict_btn.config(bg=btn_color)

# Predict button
predict_btn = tk.Button(root, text="Predict", command=predict,
                        bg=btn_color, fg="white", font=("Arial", 11, "bold"),
                        relief="flat", padx=10, pady=5)
predict_btn.bind("<Enter>", on_enter)
predict_btn.bind("<Leave>", on_leave)
predict_btn.grid(row=5, columnspan=2, pady=20)

# Start GUI loop
root.mainloop()
```
